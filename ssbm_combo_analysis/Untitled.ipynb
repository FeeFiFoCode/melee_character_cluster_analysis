```python
from PIL import Image
import os, glob
```

```python
folder_path = "ssbm_stock_icons"
file_list = []

for file_name in glob.glob(os.path.join(folder_path,"*.webp")):
    print ( "Debug 0:",file_name)
    #file_path = "{}/{}".format(folder_path,file_name)
    
    # Strip the ".webp" extension and add ".png"
    print ( "Debug 1:")
    
    ext = "png"
    new_name = "{}.{}".format(file_name[:-5],ext)
    
    # Convert the .webp file to a .png
    print ( "Debug 2:",new_name)
    
    im = Image.open(file_name).convert("RGB")
    im.save(new_name,ext)
    
    # Store a list of all the file names, can sort later
    print ( "Debug 3")
    file_list.append(new_name)
    
```
